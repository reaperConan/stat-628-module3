```python
import pandas as pd 
from functools import reduce
import string
from spellchecker import SpellChecker
spell = SpellChecker()
import nltk
porter = nltk.stem.PorterStemmer()
```

```python
user = pd.read_csv("data/c_user.csv")
tip = pd.read_csv("data/c_tip.csv")
review = pd.read_csv("data/c_review.csv")
business = pd.read_csv("data/cinema.csv")
```

```python
def append(x,y):
    return x+y
def word_process(s):
    s = s.translate(str.maketrans('', '', string.punctuation)).lower()
    s = spell.correction(s)
    return porter.stem(s)
```

```python
stop_words = pd.read_table("data/stop_words.txt")
stop_words = list(stop_words["words"])
```

```python
review
```

```python
word_init = list(map(lambda x: x.split(' '), list(review["text"])))
```

```python
words = list(reduce(append, word_init))
```

```python
word_clean = list(filter(lambda x: False if x in stop_words or x == '' else True, words))
```

```python
word = list(map(word_process, word_clean))
```
