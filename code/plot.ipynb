```python
import pandas as pd 
import numpy as np
import csv
%matplotlib inline
import matplotlib
```

```python
count0 = pd.read_csv("word_count/output0.csv", header=None)
count1 = pd.read_csv("word_count/output1.csv", header=None)
count2 = pd.read_csv("word_count/output2.csv", header=None)
count3 = pd.read_csv("word_count/output3.csv", header=None)
count4 = pd.read_csv("word_count/output4.csv", header=None)
count5 = pd.read_csv("word_count/output5.csv", header=None)
tip = pd.read_csv("word_count/tip_count.csv", header=None)

count0_weight = pd.read_csv("word_count/output_weight0.csv", header=None)
count1_weight = pd.read_csv("word_count/output_weight1.csv", header=None)
count2_weight = pd.read_csv("word_count/output_weight2.csv", header=None)
count3_weight = pd.read_csv("word_count/output_weight3.csv", header=None)
count4_weight = pd.read_csv("word_count/output_weight4.csv", header=None)
count5_weight = pd.read_csv("word_count/output_weight5.csv", header=None)
tip_weight = pd.read_csv("word_count/tip_count_weight.csv", header=None)
```

```python
word_count = pd.concat([count0,count1, count2, count3, count4, count5, tip], axis =1)
word_count.columns = ["word0","review","word1", "1 star", "word2", "2 star", "word3", "3 star", "word4", "4 star", "word5", "5 star", "wordt", "tip"]
#word_count[-50:]
```

```python
word_count = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(count0,count1, on=0), count2, on=0), count3, on=0),
                                        count4, on=0), count5, on=0), tip, on=0)
word_count.columns = ["word","review", "1 star", "2 star", "3 star", "4 star", "5 star", "tip"]
word_count.set_index('word', inplace=True)
word_count['review'] = word_count['review']/len(count0)
word_count['1 star'] = word_count['1 star']/len(count1)
word_count['2 star'] = word_count['2 star']/len(count2)
word_count['3 star'] = word_count['3 star']/len(count3)
word_count['4 star'] = word_count['4 star']/len(count4)
word_count['5 star'] = word_count['5 star']/len(count5)
word_count['tip'] = word_count['tip']/len(tip)
word_count[:50]
```

```python
word_count_weight = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(count0_weight,count1_weight, on=0), count2_weight, on=0), count3_weight, on=0),
                                        count4_weight, on=0), count5_weight, on=0), tip_weight, on=0)
word_count_weight.columns = ["word","review", "1 star", "2 star", "3 star", "4 star", "5 star", "tip"]
word_count_weight.set_index('word', inplace=True)
word_count_weight['review'] = word_count_weight['review']/len(count0_weight)
word_count_weight['1 star'] = word_count_weight['1 star']/len(count1_weight)
word_count_weight['2 star'] = word_count_weight['2 star']/len(count2_weight)
word_count_weight['3 star'] = word_count_weight['3 star']/len(count3_weight)
word_count_weight['4 star'] = word_count_weight['4 star']/len(count4_weight)
word_count_weight['5 star'] = word_count_weight['5 star']/len(count5_weight)
word_count_weight['tip'] = word_count_weight['tip']/len(tip_weight)
word_count_weight[:50]
```

```python
def variance(ty, word):
    data = ty.loc[word][1:-1]
    return np.var(data/(sum(data**2))**0.5)*1000
def word_bar(word):
    word_count.loc[word][1:-1].plot.bar(title = word+' variance=' + str(variance(word_count, word)))
def word_weight_bar(word):
    word_count_weight.loc[word][1:-1].plot.bar(title = word+' variance=' + str(variance(word_count_weight, word)))
```
